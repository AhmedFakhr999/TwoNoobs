```python
duplicates = df.duplicated()
print(f"\nNumber of duplicate rows: {duplicates.sum()}")
df = df.drop_duplicates()
print("Duplicates removed if any.")
```

```python
def standardize_country_name(name):

    return name.strip().title()

df['Country_name'] = df['Country_name'].apply(standardize_country_name)
```

```python
from sklearn.preprocessing import MinMaxScaler

features = ['Est_gdp_in_billions', 'Est_population_in_millions', 'Land_area']


scaler = MinMaxScaler()
scaler.fit(df[features])

# Inspect the internal state of the scaler
print("Scaler data_min_:")
print(scaler.data_min_)
print("\nScaler data_max_:")
print(scaler.data_max_)
print("\nScaler data_range_ (max - min):")
print(scaler.data_range_)

# Check if any of the computed ranges are zero
if (scaler.data_range_ == 0).any():
    problematic_feature = features[(scaler.data_range_ == 0).argmax()]
    print(f"\nWARNING: The scaler detected a range of ZERO for feature '{problematic_feature}'.")
else:
    print("\nAll ranges are non-zero. Proceeding with transform.")


df_scaled = pd.DataFrame(scaler.transform(df[features]), columns=features)

# Check for NaNs in the scaled data
if df_scaled.isnull().any().any():
    print("\nNaNs found in scaled data. Check the scaler's data_range_ above.")
else:
    print("\nNo NaNs in scaled data. Transform successful.")
```

```python
weights = {
    'Est_gdp_in_billions': 0.6,
    'Est_population_in_millions': 0.2,
    'Land_area': 0.2
}

df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)

df_scaled = df_scaled.set_axis(df.index, axis=0)  #


df['Development_Score'] = (
    df_scaled['Est_gdp_in_billions'] * weights['Est_gdp_in_billions'] +
    df_scaled['Est_population_in_millions'] * weights['Est_population_in_millions'] +
    df_scaled['Land_area'] * weights['Land_area']
)


print("\nDataFrame with Development Score:")

df[['Country_name', 'Est_gdp_in_billions', 'Est_population_in_millions', 'Land_area', 'Development_Score']].head()
```

### 3.1

```python
most_recent_year = df['Time_period'].max()
df_recent = df[df['Time_period'] == most_recent_year].copy()


df_recent['Income_Group'] = pd.qcut(
    df_recent['Development_Score'], 
    q=3, 
    labels=['Low Development', 'Medium Development', 'High Development']
)

print("Group sizes:")
print(df_recent['Income_Group'].value_counts().sort_index())
```

```python

indicator = 'Pct_managed_sanitation_services'


df_anova = df_recent[['Income_Group', indicator]].dropna()

print(f"\nANOVA analysis for '{indicator}' across development groups:")
```

```python

groups = df_anova.groupby('Income_Group')[indicator].apply(list)


F_statistic, p_value = stats.f_oneway(*groups)

print(f"F-statistic: {F_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

#
alpha = 0.05
if p_value < alpha:
    print(f"Reject the null hypothesis: There is a significant difference in mean {indicator} between development groups.")
else:
    print(f"Fail to reject the null hypothesis: No significant difference in mean {indicator} between groups.")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_anova, x='Income_Group', y=indicator, palette='Set2')
plt.title(f'Distribution of {indicator} by Development Group')
plt.ylabel(indicator)
plt.xlabel('Development Group')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

The ANOVA test resulted in an F-statistic of 0.1482 and a p-value of 0.8625. Since the p-value is much greater than the standard significance level ($ \alpha $ = 0.05), we fail to reject the null hypothesis. This means there is no statistically significant difference in the mean percentage of managed sanitation services across the "Low," "Medium," and "High Development" groups.

#### Interpretation and Real-World Implications:

1. Lack of Correlation Between Development Score and Sanitation Access
2. Possible Explanations:
   * The score may not capture the aspects of development most relevant to public health infrastructure. For example, it doesn't account for government effectiveness, public spending, or urbanization rates, which are critical for sanitation systems.
   * International efforts (e.g., UN Sustainable Development Goal 6) may have led to widespread improvements in sanitation, even in lower-development countries, reducing the gap between groups.

3. Policy Implication:
   * The absence of a significant difference does not mean sanitation access is adequate everywhere. Some countries in all groups might still have very low access. Policymakers should look beyond broad development metrics and analyze country-specific barriers.
   * This result suggests that current development indicators might not fully reflect quality-of-life outcomes. A more holistic index, including health, education, and infrastructure, may be needed

### 3.3 

The goal of this analysis was to develop and evaluate a **Development Score** as a composite metric to assess and compare countries based on key economic and demographic indicators. The dataset included variables such as **GDP (in billions)**, **population (in millions)**, **land area**, **managed drinking water and sanitation services**, and **regional classifications** across multiple time periods. This score was created using a weighted combination of scaled GDP, population, and land area, enabling a multidimensional view of national development.

Two significant insights emerged from the analysis. First, **population growth trends** revealed stark regional disparities, with **Sub-Saharan Africa and South Asia** experiencing rapid increases, while growth in **Europe and Oceania** has largely plateaued—highlighting shifting global demographic centers. Second, a surprising lack of correlation was found between the Development Score and access to **managed sanitation services** (ANOVA p-value = 0.86), indicating that larger or wealthier nations do not necessarily provide better basic infrastructure, likely due to the influence of targeted public health policies and international development efforts.

A key limitation of this analysis is the **potential incompleteness and definition of data**, particularly for newer or less-reported indicators like "managed" sanitation, which may not be uniformly measured across countries. Additionally, the Development Score is based on aggregate national metrics and does not account for **income inequality, governance quality, or regional disparities within countries**. For a deeper analysis, the next step would be to build a **predictive model**, such as a regression or machine learning model that incorporates social, political, and environmental variables to better explain outcomes like sanitation access or health indicators, moving beyond descriptive analysis to actionable foresight.
